```python
winner=data['winner'].value_counts()
winner.index
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=winner.index[:7],y=data['winner'].value_counts()[:7],palette='copper')
plt.xticks(rotation=45)
plt.show()
```
